### 2. Add Two Numbers

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example 1:

Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
Example 2:

Input: l1 = [0], l2 = [0]
Output: [0]
Example 3:

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 
Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.

```python
from typing import Optional, List 

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_two_numbers(l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    sum_node = ListNode(0)

    temp1 = l1 
    temp2 = l2 
    temp3 = sum_node
    carry = 0 

    while (temp1 or temp2): 
        digit_sum = 0 

        if (temp1):
            digit_sum += temp1.val 
            temp1 = temp1.next 

        if (temp2):
            digit_sum += temp2.val
            temp2 = temp2.next

        digit_sum += carry

        if (digit_sum > 9): 
            digit_sum = digit_sum % 10
            carry = 1 
        else:
            carry = 0 
            
        temp3.next = ListNode(digit_sum) 
        temp3 = temp3.next 

    if (carry == 1):
        temp3.next = ListNode(carry) 

    return sum_node.next 


def transform_arr_to_list(nums: List) -> Optional[ListNode]: 
    node = ListNode(0)
    temp = node 
    for num in nums: 
        temp.next = ListNode(num) 
        temp = temp.next 
    return node.next 

def print_linked_list(node: Optional[ListNode]) -> List[int]:
    l = [] 
    temp = node 
    while (temp):
        print(temp.val)
        l.append(temp.val) 
        temp = temp.next 
    return l


# l1 = ListNode(2)
# l1.next = ListNode(4)
# l1.next.next = ListNode(3)

# l2 = ListNode(5)
# l2.next = ListNode(6)
# l2.next.next = ListNode(4)

# digit_sum = add_two_numbers(l1, l2)

l1 = transform_arr_to_list([9,9,9,9,9,9,9])
l2 = transform_arr_to_list([9,9,9,9])

digit_sum = add_two_numbers(l1, l2) 

arr = print_linked_list(digit_sum)
print(arr)
```

### 19. Remove Nth Node From End of List

Hint
Given the head of a linked list, remove the nth node from the end of the list and return its head.

Example 1:

Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
Example 2:

Input: head = [1], n = 1
Output: []
Example 3:

Input: head = [1,2], n = 1
Output: [1]

```python
from typing import Optional, List 

class ListNode:
    def __init__(self, val=0, next=None) -> None: 
        self.val = val 
        self.next = next 

def transform_from_array_to_linked_list(nums: List[int]) -> ListNode:
    head = ListNode(0)
    temp = head  
    for num in nums: 
        temp.next = ListNode(num) 
        temp = temp.next 
    return head.next

def print_list(head: Optional[ListNode]) -> None: 
    temp = head 
    while (temp):
        print(temp.val) 
        temp = temp.next 
        
def remove_nth_from_end(head: ListNode, n: int) -> ListNode:
    stack = [] 
    head = ListNode(-1, head) 
    temp = head 

    while (temp):
        stack.append(temp) 
        temp = temp.next 

    tail_node = None
    current_node = None

    i = 1 
    while (stack != []): 
        current_node = stack.pop() 

        if (i == n):
            del current_node 
        else: 
            current_node.next = tail_node 
            tail_node = current_node

        i = i + 1 

    return head.next

head = transform_from_array_to_linked_list([1])
res = remove_nth_from_end(head, 1) 
print_list(res) 
```

### 24. Swap Nodes in Pairs

Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)

Example 1:

Input: head = [1,2,3,4]
Output: [2,1,4,3]
Example 2:

Input: head = []
Output: []
Example 3:

Input: head = [1]
Output: [1]
 

Constraints:

The number of nodes in the list is in the range [0, 100].
0 <= Node.val <= 100

```python
from typing import Optional, List 

class ListNode:
    def __init__(self, val=0, next=None) -> None: 
        self.val = val 
        self.next = next 

def transform_from_array_to_linked_list(nums: List[int]) -> Optional[ListNode]: 
    head = ListNode(0)
    temp = head  
    for num in nums: 
        temp.next = ListNode(num) 
        temp = temp.next 
    return head.next

def print_list(head: ListNode) -> None: 
    temp = head 
    while (temp):
        print(temp.val) 
        temp = temp.next 

def swap_pairs(head: Optional[ListNode]) -> Optional[ListNode]: 
    if (head and head.next == None):
        return head 

    new_head = ListNode(-1)
    
    current_head_temp = new_head   
    temp = head 

    while (temp): 
        if (temp.next):
            node1 = temp
            node2 = temp.next 
            node1.next = node2.next 
            node2.next = node1 
            current_head_temp.next = node2 
            current_head_temp = node1
            temp = node1.next 
        else: 
            temp = temp.next 

    return new_head.next

head = transform_from_array_to_linked_list([1,2,3,4])
res = swap_pairs(head)
print_list(res)
```
